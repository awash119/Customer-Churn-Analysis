```python
from sklearn.preprocessing import OneHotEncoder
import pandas as pd
import numpy as np 
```

```python
df=pd.read_csv("data/raw/ChurnModelling_Raw_1.csv")
df.head()
```

```python
df_encord=df[["Geography","Gender"]]
encoder=OneHotEncoder(drop="first",sparse_output=False)
```

```python

encoded_df=encoder.fit_transform(df_encord[["Geography","Gender"]])
```

```python
encoded_df=pd.DataFrame(encoded_df,columns=encoder.get_feature_names_out(["Geography", "Gender"]))
encoded_df.head()
```

```python
df=pd.concat(df,encoded_df,axis=1)
```
