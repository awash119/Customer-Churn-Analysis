```python
from sklearn.preprocessing import OneHotEncoder
import pandas as pd
import numpy as np 
```

```python
df=pd.read_csv("data/raw/ChurnModelling_Raw_1.csv")
df.head()
```

```python
df_encord=df[["Geography","Gender"]]
encoder=OneHotEncoder(drop="first",sparse_output=False)
```

```python

encoded_df=encoder.fit_transform(df_encord[["Geography","Gender"]])
```

```python
encoded_df=pd.DataFrame(encoded_df,columns=encoder.get_feature_names_out(["Geography", "Gender"]))
encoded_df.head()
```

```python
df=pd.concat([df,encoded_df],axis=1)
```

```python
df
```

```python
df.drop(columns=["Geography","Gender"],inplace=True)
from sklearn.preprocessing import StandardScaler
```

```python
def scale(df,columns):
    scaler=StandardScaler()
    df_col_to_scale=df[columns]
    df=df.drop(columns=columns,axis=1)
    df_col_to_scale=scaler.fit_transform(df_col_to_scale[columns])
    df_col_to_scale=pd.DataFrame(df_col_to_scale,columns=columns,index=df.index)
    df=pd.concat([df,df_col_to_scale],axis=1)
    print(df.head())

columns=["Age","Balance", "EstimatedSalary"]
```

```python
scale(df,columns)
```
